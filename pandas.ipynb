```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_excel('mockdata.xlsx', 'custsales')
df = df.groupby('orderno')['dat', 'city', 'linetotal'].agg({'dat': min, 'city': min, 'linetotal': sum})
                                                            
df.head()
```

```python
plt.style.use('ggplot')

df.groupby('city').hist()
```

```python
df = pd.DataFrame({'A' : ['foo', 'bar', 'foo', 'bar', 'foo', 'bar', 'foo', 'foo'],
                   'B' : ['one', 'one', 'two', 'three', 'two', 'two', 'one', 'three'],
                   'C' : np.random.randn(8),
                   'D' : np.random.randn(8)})

print(df)

def get_letter_type(letter):
    if letter.lower() in 'aeiou':
        return 'vowel'
    else:
        return 'consonant'

grouped = df.groupby(get_letter_type, axis=1)
grouped.last()
```
